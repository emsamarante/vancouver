```python
import pandas as pd
```

```python
data = pd.read_csv("../data/dataset.csv", index_col=0)
data.head()
```

```python
data.tail()
```

```python
aux = data[['TYPE', 'YEAR', 'MONTH', 'MINUTE']].groupby(by=['TYPE', 'YEAR', 'MONTH']).count().reset_index().rename(columns={'MINUTE':'CONTAGEM'})
```

```python
aux2 = aux[(aux['TYPE'].isin(['Break and Enter Commercial'])) & (aux['MONTH'].isin([1]))]
aux2
```

```python
aux2.loc[:,'cont_py'] = aux2.CONTAGEM.shift(periods=1)
aux2
```

```python
aux2.loc[:,'relative'] = round(100*(aux2['CONTAGEM'] - aux2['cont_py'])/aux2['cont_py'], 2)
```
